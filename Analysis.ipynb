```python
import pandas as pd
import pandasql as ps
```

```python
customers_data = pd.read_csv('customer.csv')
products_data = pd.read_csv('products.csv')
sales_data = pd.read_csv('sales.csv')
dealerships_data = pd.read_csv('dealerships.csv')
```

#### List of all emails of customers in the state of Flourida

```python
ps.sqldf("""
    SELECT email from customers_data as df 
    WHERE state = 'FL' 
    ORDER BY email
    """)
```

#### Names of all customers in NY

```python
ps.sqldf("""
    SELECT first_name, last_name 
    FROM customers_data as df
    WHERE CITY ='New York City' AND state ='NY' 
    ORDER BY last_name, first_name
    """)
```

#### List of all customers whith no null phone numbers

```python
ps.sqldf("""
    SELECT df.* 
    From customers_data as df 
    WHERE phone IS NOT NULL 
    ORDER BY date_added 
    LIMIT 5
    """)
```

#### List all customers with high savings in ny state

```python
ps.sqldf("""
SELECT 
c.first_name, c.last_name, c.phone, c.city, c.state,
p.model, p.product_type,p.base_msrp,
s.sales_amount, s.channel,
COALESCE(s.dealership_id, 0) as dealership_yes,
CASE WHEN p.base_msrp - s.sales_amount >1000 THEN 1 ELSE 0 END AS high_savings
FROM sales_data s
INNER JOIN customers_data c ON c.customer_id = s.customer_id 
INNER JOIN products_data p ON p.product_id = s.product_id
LEFT JOIN dealerships_data d ON s.dealership_id = d.dealership_id
WHERE c.state = 'NY' and high_savings = 1
ORDER BY c.first_name
""")
```

#### Total Sales done by each states

```python
ps.sqldf("""
SELECT c.state as state, SUM(sales_amount) as total_sales
FROM sales_data s
INNER JOIN customers_data c
ON c.customer_id = s.customer_id
GROUP BY state
Order BY state
""")
```

```python
ps.sqldf("""

SELECT s.dealership_id, SUM(s.sales_amount) as total_number
FROM sales_data s
INNER JOIN dealerships_data d
ON d.dealership_id = s.dealership_id
GROUP BY d.dealership_id
ORDER BY total_number DESC
LIMIT 5
""")
```

```python
ps.sqldf("""

SELECT s.dealership_id, COUNT(*)
FROM sales_data s
WHERE channel='dealership'
GROUP BY 1
ORDER BY 2 DESC
LIMIT 5

""")
```

```python
ps.sqldf("""
SELECT s.channel, s.product_id, AVG(sales_amount) as avg_sales_amount
FROM sales_data s
GROUP BY 
s.channel, s.product_id
ORDER BY s.channel, s.product_id
""")
```
